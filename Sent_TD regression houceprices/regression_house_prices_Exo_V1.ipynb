```python
#creat the NN network in a Regressor Class
class Regressor(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(332,256)
        self.fc1 = nn.Linear(144,72)
        self.fc1 = nn.Linear(72, 18)
        self.fc1 = nn.Linear(18, 1)
        
    def forward(self,x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.relu(self.fc4(x))
        
        return x
```

```python
#set the training and test batch into 50 bathes
#You can use numpy.split to split along the first axis n times, 
#where n is the number of desired batches. Thus, the implementation would look like this -

train_batch = np.array_split(X_train, 50)
label_batch = np.array_split(y_train, 50)
```

```python
# creat the torch tensors for  train batch and  labels (y train values)
for i in range(len(train_batch)):
    train_batch[i] = torch.from_numpy(train_batch[i].values).float()
for i in range(len(label_batch)):
    label_batch[i] = torch.from_numpy(label_batch[i].values).float().view(-1, 1) # view is added so that sizes are respected
#it can be verified that now train test batches are tensors
```

```python
model = Regressor()

#initialize the weights
model.apply(weights_init)

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

epochs = 300

train_losses, test_losses = [], []
for e in range(epochs):
    model.train() #just setting the mode for training
    train_loss = 0
    for i in range(len(train_batch)):
        #clear gradients
        optimizer.zero_grad()
        
        #gather output
        output = model(train_batch[i])
        
        #calcluate loss
        loss = torch.sqrt(criterion(torch.log(output), torch.log(label_batch[i]))) #alternative
        # loss = criterion(torch.log(output), torch.log(label_batch[i]))
        
        #backprop
        loss.backward()
        
        #take a stpe of optimizer
        optimizer.step()
        
        #add loss to train
        running_loss += loss.item()
        
    else:
        test_loss = 0
        accuracy = 0
        
# with evaluation mode on , get the 0utput from the network using test data
# and get the test loss. 

        with torch.no_grad():
            model.eval()
            predictions = model(X_val)
            test_loss += criterion(torch.log(predictions), torch.log(y_val))
            
                
        train_losses.append(train_loss/len(train_batch))
        test_losses.append(test_loss)

        print("Epoch: {}/{}.. ".format(e+1, epochs),
              "Training Loss: {:.3f}.. ".format(train_loss/len(train_batch)),
              "Test Loss: {:.3f}.. ".format(test_loss))
```
